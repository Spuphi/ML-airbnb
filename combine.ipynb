```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
listings_clean = pd.read_csv("data/listings_clean.csv")
calender_features = pd.read_csv("data/calender_features.csv")
```

```python
combined = listings_clean.merge(calender_features, left_on="id", right_on="listing_id", how="left")
combined = combined.drop(columns=["listing_id", "id_y", "price_y", "availability_365_y"])
```

```python
combined = combined.rename(columns={
    "price_x": "price",
    "id_x": "id",
    "availability_365_x": "availability_365"
})

combined.head()
print(combined["availability_365"])
```

```python
combined.to_csv("data/combined_data.csv", index=False)
```
