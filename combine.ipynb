```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
listings_clean = pd.read_csv("data/listings_clean.csv")
calender_features = pd.read_csv("data/calender_features.csv")
review_sentiment = pd.read_csv("data/listing_sentiment_multilingual.csv")
spatial = pd.read_csv("data/listings_spatial.csv")
```

```python
combined = listings_clean.merge(calender_features, left_on="id", right_on="listing_id", how="left")
combined = combined.drop(columns=["listing_id", "id_y", "price_y", "availability_365_y"])

combined = combined.merge(review_sentiment, how="left", left_on= "id_x", right_on="listing_id")
combined = combined.drop(columns=["listing_id"])

combined = combined.merge(spatial.drop(columns=["price", "latitude", "longitude"]), how="left", left_on= "id_x", right_on="id")
combined = combined.drop(columns=["id"])
```

```python
combined = combined.rename(columns={
    "price_x": "price",
    "id_x": "id",
    "availability_365_x": "availability_365"
})

print(combined.shape)
display(combined.head())
```

```python
combined.to_csv("data/combined_data.csv", index=False)
```
